```python
# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')  #tracks dataset
df_tracks.head()
```

```python
# Checking for null values in the dataset

df_tracks.isnull().sum()
```

```python
# Getting info of the dataset

df_tracks.info()
```

```python
# 10 least popular songs in the dataset

sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
df_tracks.describe()
```

```python
df_tracks.describe().transpose()
```

```python
# top10 most popular songs

most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
# make release date as index for every row

df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# to check artist at a particular location

df_tracks[["artists"]].iloc[25] #artist at 25th location
```

```python
# convert duration of song into seconds from ms

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)
df_tracks.head()
```

```python
#Creating a correlation table

corr_df = df_tracks.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
corr_df
```

```python
# Correlation plot

plt.figure(figsize= (14,6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap='inferno', linewidths = 1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90);
```

```python
# as we observe co-relation b/w loudness and energy, let's take some sample data and build a regression plot

sample_df = df_tracks.sample(int(0.005*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
# Now let's build a regression plot for Loudness and Energy relation

plt.figure(figsize= (10,6))
sns.regplot(data = sample_df, y='loudness', x = 'energy', scatter_kws={'color': "cyan"}, line_kws={'color': 'blue'}).set(title="Loudness vs Energy Correlation")
```

```python
# Now let's build another regression plot for Popularity and Acousticness relation


plt.figure(figsize= (10,6))
sns.regplot(data = sample_df, y='popularity', x = 'acousticness', scatter_kws={'color': 'blue'}, line_kws={'color': 'red'}).set(title="Popularity vs Acousticness Correlation")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
df_tracks.head()
```

```python
# Plot for no.of songs released every year 
    
sns.displot(years, discrete=True, aspect = 2, height = 5, kind = 'hist').set(title='Number of Songs per year');
```

```python
# Duration of songs over years

total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x = years, y = total_dr, ax = ax, errwidth = False).set(title = "Year vs Duration")
plt.xticks(rotation = 90);
```

```python
df_genre = pd.read_csv("SpotifyFeatures.csv")
df_genre.head()
```

```python
# Bar plot to represent avg duration of songs in each Genre

plt.title("Duration of songs in different Genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y='genre', x = 'duration_ms', data= df_genre)
plt.xlabel("Duration in milli seconds")
plt.ylabel("Genres");
```

```python
# get top 5 genres by popularity

sns.set_style(style = 'darkgrid')
plt.figure(figsize=(10,5))
famous = df_genre.sort_values('popularity', ascending = False).head(10) # here repetitions are present, so wetook 10 instead of 5
sns.barplot(y='genre', x='popularity', data=famous).set(title='Top 5 Genres by popularity');
```
